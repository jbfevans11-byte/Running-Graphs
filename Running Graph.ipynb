```python
import numpy as np

import matplotlib.pyplot as plt

import scipy.optimize as scpo 

import pandas as pd

df = pd.read_excel('Gym.xlsx', sheet_name = 'Running')

df = df.values

Day = df[37:1000,2]
Time = df[37:1000, 3]

Day = pd.to_numeric(Day, errors="coerce")
Time = pd.to_numeric(Time, errors="coerce")

Time[Time == 0] = np.nan #Replacing the 0's

mask = ~np.isnan(Time)
Day_clean = Day[mask]
Time_clean = Time[mask]

def Line1(x, gradient1, intercept1): 
    return gradient1*x + intercept1 

actual_fit_parameters, covariance_matrix = scpo.curve_fit(Line1, Day_clean, Time_clean) 

fit_gradient1, fit_intercept1 = actual_fit_parameters
ybestfit1 = Line1(Day, fit_gradient1, fit_intercept1)

Day_extended = np.linspace(1, 250)

ybestfit_extended = Line1(Day_extended, fit_gradient1, fit_intercept1)

plt.scatter(Day, Time, color="blue", label="Data Points")

plt.plot(Day_extended, ybestfit_extended, color="red", label="Extrapolated Best Fit")


plt.gca().set_xlim([-0.5,250])

plt.gca().set_ylim([18,31])

plt.axhline(20, color="black", linestyle="--")

plt.xlabel("Day")
plt.ylabel("Time")
plt.show()

equation = f"y = {fit_gradient1:.3f}x + {fit_intercept1:.3f}"
print(equation)

def Time(y):
    return (y-fit_intercept1)/(fit_gradient1)

print(Time(20))
```
